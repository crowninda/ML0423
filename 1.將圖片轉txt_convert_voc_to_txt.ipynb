```python
import xml.etree.ElementTree as ET
import os
from os import getcwd
from glob import glob

classes = ['01_baote_bottle', '02_plastic_bottle_cap', '03_straw', '04_take_out_drink_cup', '05_disposable_tableware', '06_Plastic_bag', '07_glass_bottle', '08_float', '09_styrofoan', '10_other_garbage' ]

list_file = open(os.path.join('drive/My Drive', 'anno_beach_0714.txt'), 'w') # target .txt file

for path in sorted(glob(os.path.join(FOLDER_PATH, '第一次影像蒐集/海廢lable/lable/*.xml'))):
    file_id = ''.join(path.split('/')[-1].split('.')[:-1])
    print(file_id)
    in_file = open(path)
    tree=ET.parse(in_file)
    root = tree.getroot()
    

    list_file.write(f'第一次影像蒐集/海廢lable/image/{file_id}.jpg')
    for obj in root.iter('object'):
        cls = obj.find('name').text
        cls_id = classes.index(cls)
        xmlbox = obj.find('bndbox')
        b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
        list_file.write(" " + ",".join([str(a) for a in b]) + ',' + str(cls_id))
    list_file.write('\n')
list_file.close()
```

```python
import xml.etree.ElementTree as ET
import os
from os import getcwd
from glob import glob

classes = ['01_baote_bottle', '02_plastic_bottle_cap', '03_straw', '04_take_out_drink_cup', '05_disposable_tableware', '06_Plastic_bag', '07_glass_bottle', '08_float', '09_styrofoan', '10_other_garbage' ]

list_file = open(os.path.join('drive/My Drive', 'anno_beach_0714.txt'), 'w') # target .txt file
# .[jJ][pP][gG]
for path in sorted(glob(os.path.join(FOLDER_PATH, '第一次影像蒐集/海廢lable/image/*'))):
    file_id = ''.join(path.split('/')[-1].split('.')[:-1])
    file_id_jpg = ''.join(path.split('/')[-1].split('.')[-1])
    print(file_id)
    path_1 = os.path.join(f'/content/drive/My Drive/yolo/第一次影像蒐集/海廢lable/lable/{file_id}.xml')
    print(path_1)
    in_file = open(path_1)
    tree=ET.parse(in_file)
    root = tree.getroot()
    

    list_file.write(f'第一次影像蒐集/海廢lable/image/{file_id}.{file_id_jpg}')
    for obj in root.iter('object'):
        cls = obj.find('name').text
        cls_id = classes.index(cls)
        xmlbox = obj.find('bndbox')
        b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
        list_file.write(" " + ",".join([str(a) for a in b]) + ',' + str(cls_id))
    list_file.write('\n')
list_file.close()
```
